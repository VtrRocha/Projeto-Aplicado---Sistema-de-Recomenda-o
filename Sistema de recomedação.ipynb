```python
import pandas as pd 
import numpy as np
```

```python
filmes = pd.read_csv('movies_metadata.csv', low_memory=False)
filmes.head()
```

```python
avaliacoes = pd.read_csv('ratings.csv', low_memory=False)
avaliacoes.head()
```

```python
#Selecionando somente as colunas nescessarias e as renomeando
filmes = filmes[['id','original_title','original_language','vote_count']]


filmes.rename(columns = {'id':'ID_FILME','original_title':'Titulo','original_language':'Linguagem','vote_count':'Qtd_avaliacoes'},inplace=True)

filmes.head()
```

```python
avaliacoes = avaliacoes[['userId','movieId','rating']]

avaliacoes.rename(columns= {'userId': 'Id_Usuario','movieId':'ID_FILME','rating':'Avaliacao'}, inplace=True)

avaliacoes.head()
```

```python
#Verificando se há valores nulos
filmes.isna().sum()
```

```python
#Removendo os valores nulos para não impactar no modelo porque são poucos valores
filmes.dropna(inplace=True)
filmes.isna().sum()
```

```python
avaliacoes.isna().sum()
```

```python
#Gerando uma contagem de avaliações por usuario e considerando usuarios que avaliaram mais de 999 vezes

avaliacoes['Id_Usuario'].value_counts()
qtd_avaliacoes = avaliacoes['Id_Usuario'].value_counts() >999
y = qtd_avaliacoes[qtd_avaliacoes].index
y.shape
```

```python
#visualizando o tamanho do dataset "Avaiacoes"
avaliacoes.shape
```

```python
#Pegando os usuarios que avaliram mais de 999 vezes
avaliacoes =  avaliacoes[avaliacoes['Id_Usuario'].isin(y)]
avaliacoes.shape
```

```python
avaliacoes.head
```

```python
#gerando os filmes que tiveram mais de 999 avaliações
filmes = filmes[filmes['Qtd_avaliacoes']>999]
```

```python
#Agrupando e visualiando a quantidade de filmes pela linguagem
filmes_linguagem = filmes['Linguagem'].value_counts()
filmes_linguagem.head(20)
```

```python
filmes = filmes[filmes['Linguagem'] == 'en']
filmes.info()
```

```python
#Convertedo o Id_Filme para interiro
filmes['ID_FILME'] = filmes['ID_FILME'].astype(int)
```

```python
filmes.info()
```

```python
#Conectando os dataframes

avaliacoes_e_filmes = avaliacoes.merge(filmes, on='ID_FILME')
avaliacoes_e_filmes.head()
```

```python
#visualizando o dataframe unificado
avaliacoes_e_filmes.head(20)
```

```python
#removendo valores duplicados do df e visualizando se tiveram alterações na quantidade de registros
avaliacoes_e_filmes.drop_duplicates(['Id_Usuario','ID_FILME'],inplace=True)
avaliacoes_e_filmes.shape
```

```python
#removendo a coluna ID filmes 
del avaliacoes_e_filmes['ID_FILME']
```

```python
#fazendo uma pivot table para fazer com que cada usuario vire uma variavel com as respectivas notas
filmes_pivot = avaliacoes_e_filmes.pivot_table(columns=  'Id_Usuario',index = 'Titulo', values='Avaliacao')
filmes_pivot.head(20)
```

```python
#tranformando os valores nulos para 0
filmes_pivot.fillna(0,inplace=True)
filmes_pivot.head()
```

```python
#Importando o pacote csr_matrix do pacote scipy
#Metodo para criarmos uma matrix sparsa
from scipy.sparse import csr_matrix
filmes_sparse = csr_matrix(filmes_pivot)
```

```python
#Importando o algoritmo KNN do Sckit Learn
from sklearn.neighbors import NearestNeighbors
```

```python
#Criando e treinando o modelo
modelo = NearestNeighbors(algorithm='brute')
modelo.fit(filmes_sparse)
```

#PREVISÕES DE FILMES

```python
#127 Hours
distances,sugestions = modelo.kneighbors(filmes_pivot.filter(items=['127 Hours'],axis=0).values.reshape(1,-1))

for i in range(len(sugestions)):
    print(filmes_pivot.index[sugestions[i]])
```
